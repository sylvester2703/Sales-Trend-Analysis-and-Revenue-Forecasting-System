```python
import numpy as np
import pandas as pd
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\Sales trend and revenue forecasting\cleaned_walmart_sales_v2.csv")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.drop_duplicates()
df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mean())
df['Fuel_Price'] = df['Fuel_Price'].fillna(df['Fuel_Price'].mean())
df['CPI'] = df['CPI'].fillna(df['CPI'].mean())
df['Unemployment'] = df['Unemployment'].fillna(df['Unemployment'].mean())
if 'IsHoliday_x' in df.columns:
    df.rename(columns={'IsHoliday_x': 'IsHoliday'}, inplace=True)
if 'IsHoliday_y' in df.columns:
    df.drop(columns=['IsHoliday_y'], inplace=True)
print(df.isnull().sum())
```
